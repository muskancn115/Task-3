```python
import pandas as pd   
import numpy as np   
import matplotlib.pyplot as plt    
data=pd.read_csv("C:\\Users\\k2roc\\Desktop\\DATA - 3.csv") 
data.head(10)
```

```python
#scatter plot
plt.scatter(data['gender'],data['age'])

#Title
plt.title('Scatter plot')

#Labels
plt.xlabel('age')
plt.ylabel('gender')

#show chart
plt.show()
```

```python
#Histogram
plt.hist(data['age'])

#title
plt.title('histogram')
plt.show()
```

```python
#Line plot
plt.plot(data['gender'])
plt.plot(data['age'])

#Title
plt.title('Line plot')

#Labels
plt.xlabel('age')
plt.ylabel('gender')

#show chart
plt.show()
```
